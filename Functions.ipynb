Basic Calculator

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

def display_result(a, b, operation):
    if operation == "addition":
        print(f"Addition of {a} and {b} is {add(a, b)}")
    elif operation == "subtraction":
        print(f"Subtraction of {a} and {b} is {subtract(a, b)}")
    elif operation == "multiplication":
        print(f"Multiplication of {a} and {b} is {multiply(a, b)}")
    elif operation == "division":
        print(f"Division of {a} and {b} is {divide(a, b)}")
    else:
        print("Invalid operation")

# Example usage
num1 = 10
num2 = 5

display_result(num1, num2, "addition")
display_result(num1, num2, "subtraction")
display_result(num1, num2, "multiplication")
display_result(num1, num2, "division")
```

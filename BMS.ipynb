Start Program

get userâ€™s monthly budget

Repeat until user wants to exit

show current budget status

process the userâ€™s choice

show final summary

```python
# Weekly Budget Management System for Students

# ---------- Functions ----------

def get_budget_status(budget, spent):
    remaining = budget - spent
    percentage_spent = (spent / budget) * 100

    print("\n--- Budget Status ---")
    print(f"Total Budget: ${budget}")
    print(f"Total Spent: ${spent}")
    print(f"Remaining: ${remaining}")
    print(f"Spent: {percentage_spent:.2f}%")

    if remaining >= budget * 0.5:
        print("Status: Good ðŸ‘")
    elif remaining > 0:
        print("Status: Careful âš ï¸")
    else:
        print("Status: Over Budget âŒ")

    return remaining


def can_buy(item_price, remaining):
    return item_price > 0 and item_price <= remaining


def apply_discount(price, discount_percent):
    discount_amount = (discount_percent / 100) * price
    return price - discount_amount


def show_summary(spending):
    print("\n--- Weekly Spending Summary ---")
    total = 0
    for category, amount in spending.items():
        print(f"{category}: ${amount}")
        total += amount
    print(f"Total Spent: ${total}")


# ---------- Main Program ----------

def main():
    WEEKLY_BUDGET = 100
    spent = 0

    spending = {
        "Food & Drinks": 0,
        "School Supplies": 0,
        "Entertainment": 0,
        "Transportation": 0
    }

    print("Welcome to Weekly Budget Manager ðŸ’¸")
    print("Weekly Allowance: $100")

    while True:
        remaining = get_budget_status(WEEKLY_BUDGET, spent)

        print("\nMenu:")
        print("1. Buy Item")
        print("2. View Spending Summary")
        print("3. Exit")

        choice = input("Choose an option (1-3): ")

        if choice == "1":
            print("\nCategories:")
            print("1. Food & Drinks")
            print("2. School Supplies")
            print("3. Entertainment")
            print("4. Transportation")

            cat_choice = input("Select category (1-4): ")
            categories = list(spending.keys())

            if cat_choice not in ["1", "2", "3", "4"]:
                print("Invalid category.")
                continue

            category = categories[int(cat_choice) - 1]
            price = float(input("Enter item price: $"))

            discount = float(input("Enter discount percentage (0 if none): "))
            final_price = apply_discount(price, discount)

            print(f"Final price after discount: ${final_price}")

            if can_buy(final_price, remaining):
                spending[category] += final_price
                spent += final_price
                print("Purchase successful âœ…")
            else:
                print("Cannot afford this item âŒ")

        elif choice == "2":
            show_summary(spending)

        elif choice == "3":
            print("\nExiting program...")
            show_summary(spending)
            get_budget_status(WEEKLY_BUDGET, spent)
            print("Good luck managing your money! ðŸ‘‹")
            break

        else:
            print("Invalid option. Try again.")


# ---------- Run Program ----------
main()
```
